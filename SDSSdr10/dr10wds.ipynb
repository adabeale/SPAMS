# SDSS WDs from DR10 ([Kepler et al. 2014](http://arxiv.org/abs/1411.4149))

```python
%pylab inline
import numpy as np
import matplotlib
matplotlib.rcParams['font.size'] = 15
from matplotlib import pyplot as plt

rawfile = open('dr910.final.txt').read().splitlines()

names = [] 
u = np.zeros_like(rawfile[1:], dtype=float)
g = np.zeros_like(rawfile[1:], dtype=float)
r = np.zeros_like(rawfile[1:], dtype=float)
logg = np.zeros_like(rawfile[1:], dtype=float)
logg_unc = np.zeros_like(rawfile[1:], dtype=float)
spectype = []
Teff = np.zeros_like(rawfile[1:], dtype=float)
Teff_unc = np.zeros_like(rawfile[1:], dtype=float)

for i, line in enumerate(rawfile[1:]):
    splitline = line.split(' ')
    names.append(splitline[0])
    u[i] = float(splitline[3])
    g[i] = float(splitline[5])
    r[i] = float(splitline[7])
    Teff[i] = float(splitline[14])
    Teff_unc[i] = float(splitline[15])
    logg[i] = float(splitline[16])
    logg_unc[i] = float(splitline[17])
    spectype.append(splitline[18].replace('\t',''))

# Narrow down spectral types with metal pollution (Z), without alternative 
# classifications (/) and without uncertain classifications (:)
metalpollutedinds = np.array([i for i in range(len(spectype)) if 'Z' in spectype[i].upper()\
                              and '/' not in spectype[i] and len(spectype[i]) < 10 \
                              and ':' not in spectype[i]])
```

```python
mp_names = [names[i] for i in metalpollutedinds]
mp_Teff = Teff[metalpollutedinds]
mp_Teff_unc = Teff_unc[metalpollutedinds]
mp_g = g[metalpollutedinds]

fig, ax = plt.subplots(figsize=(14,8))
ax.errorbar(mp_g, mp_Teff, yerr=mp_Teff_unc, fmt='k.')
ax.set_xlabel('$g$')
ax.set_ylabel('$T_{eff}$')
ax.set_title('All Kepler et al. (2014) Metal Polluted WDs')
plt.show()
```

```python
metalpollutedzzinds = np.array([i for i in range(len(spectype)) if 'Z' in spectype[i].upper()\
                              and '/' not in spectype[i] and len(spectype[i]) < 10 \
                              and ':' not in spectype[i] and 11100 < Teff[i] < 12500])

mp_zz_names = [names[i] for i in metalpollutedzzinds]
mp_zz_Teff = Teff[metalpollutedzzinds]
mp_zz_Teff_unc = Teff_unc[metalpollutedzzinds]
mp_zz_g = g[metalpollutedzzinds]
mp_zz_logg = logg[metalpollutedzzinds]
mp_zz_logg_unc = logg_unc[metalpollutedzzinds]
mp_zz_spectype = [spectype[i] for i in metalpollutedzzinds]
# cmap = mpl.cm.spectral
# for i in range(len(mp_zz_g)):
#     ax.errorbar(mp_zz_g[i], mp_zz_Teff[i], yerr=mp_zz_Teff_unc[i], \
#                 fmt='o', color=cmap((mp_zz_logg[i] - mp_zz_logg.min())/mp_zz_logg.max()))

fig, ax = plt.subplots(2, 1, figsize=(14,14), sharex=True)
ax[0].errorbar(mp_zz_g, mp_zz_Teff, yerr=mp_zz_Teff_unc, fmt='ko')
ax[0].set_ylabel('$T_{eff}$')
ax[0].set_title('Potential ZZ-Ceti+Metal Polluted WDs in Kepler et al. (2014)')
for lim in [11100, 12500]:
    ax[0].axhline(lim, ls='--', color='k')
    
ax[1].errorbar(mp_zz_g, mp_zz_logg, yerr=mp_zz_logg_unc, fmt='ko')
ax[1].set_xlabel('$g$')
ax[1].set_ylabel('$\log g$')
fig.subplots_adjust(hspace=0.05)
plt.show()
```
